First, let's load the data from `train_KNN.csv` and `test_KNN.csv` into pandas DataFrames.

```python
import pandas as pd

df_train = pd.read_csv('/content/train_KNN.csv')
df_test = pd.read_csv('/content/test_KNN.csv')

print("Training data head:")
display(df_train.head())
print("\nTesting data head:")
display(df_test.head())
```

Next, we'll separate the target variable ('Survived') from the features. This ensures 'Survived' is in `y` and 'Embarked' remains in `X` for further processing.

```python
y_train = df_train['Survived']
X_train = df_train.drop('Survived', axis=1)

y_test = df_test['Survived']
X_test = df_test.drop('Survived', axis=1)

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

Now, let's clean up the features by dropping columns that are not useful for our model, such as 'PassengerId', 'Name', 'Ticket', and 'Cabin'.

```python
X_train = X_train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
X_test = X_test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

print("X_train columns after dropping:", X_train.columns.tolist())
print("X_test columns after dropping:", X_test.columns.tolist())
```

Next, we'll handle missing values. For numerical columns ('Age', 'Fare'), we'll use the median. For categorical columns like 'Embarked', we'll use the mode (most frequent value).

```python
import numpy as np

# Convert categorical features into dummy variables
X_train = pd.get_dummies(X_train, columns=['Sex', 'Embarked'], drop_first=True, dtype=int)
X_test = pd.get_dummies(X_test, columns=['Sex', 'Embarked'], drop_first=True, dtype=int)

# Align columns - crucial for consistent feature sets between train and test
# Add missing columns to X_test with 0, and remove extra columns from X_test
missing_cols_train = set(X_train.columns) - set(X_test.columns)
for c in missing_cols_train:
    X_test[c] = 0

missing_cols_test = set(X_test.columns) - set(X_train.columns)
for c in missing_cols_test:
    X_test.drop(c, axis=1, inplace=True)

X_test = X_test[X_train.columns] # Ensure the order of columns is the same

print("X_train head after one-hot encoding:")
display(X_train.head())
print("\nX_test head after one-hot encoding:")
display(X_test.head())
```

Now that our data is clean and all numerical, we can scale the features using `StandardScaler`. This is an important preprocessing step for many machine learning algorithms, ensuring all features contribute equally.

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
schaler = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
scaled_X_train = schaler.fit_transform(X_train)
scaled_X_test = schaler.transform(X_test)

print("Scaled X_train (first 5 rows):\n", scaled_X_train[:5])
print("\nScaled X_test (first 5 rows):\n", scaled_X_test[:5])
```

### Decision Tree Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train the Logistic Regression model
log_reg = LogisticRegression(random_state=42)
log_reg.fit(scaled_X_train, y_train)

# Predict and evaluate
y_pred_log = log_reg.predict(scaled_X_test)
log_accuracy = accuracy_score(y_test, y_pred_log)
print(f"Logistic Regression Accuracy: {log_accuracy:.4f}")
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the Decision Tree model
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(scaled_X_train, y_train)

# Predict and evaluate
y_pred_dt = dt_clf.predict(scaled_X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")
```

### Model Performance Comparison

```python
import pandas as pd

# Create a DataFrame to compare accuracies
comparison_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree'],
    'Accuracy': [log_accuracy, dt_accuracy]
})

print("\nModel Performance Comparison:")
display(comparison_df.sort_values(by='Accuracy', ascending=False))
```
